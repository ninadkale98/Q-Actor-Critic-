```python
# Importing libraries
import gymnasium as gym
import numpy as np
from itertools import count
from collections import namedtuple
```

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
from torch.distributions import Categorical
```

```python
SavedAction = namedtuple('SavedAction', ['log_prob', 'value'])
```

```python
env = gym.make('CartPole-v1') # We make the Cartpole environment here
```

```python
# You can move either left or right to balance the pole
# Lets implement the Actor critic network
class ActorCritic(nn.Module):
    def __init__(self):
        super(ActorCritic, self).__init__()
        self.fc1 = nn.Linear(4, 128) # 4 because there are 4 parameters as the observation space
        self.actor = nn.Linear(128, 2) # 2 for the number of actions
        self.critic = nn.Linear(128, 1) # Critic is always 1
        self.saved_actions = []
        self.rewards = []
    def forward(self, x):
        x = F.relu(self.fc1(x))
        action_prob = F.softmax(self.actor(x), dim=-1)
        state_values = self.critic(x)
        return action_prob, state_values
```

```python
model = ActorCritic()
optimizer = optim.Adam(model.parameters(), lr=1e-4)
eps = np.finfo(np.float32).eps.item()
```

```python
def select_action(state):
    state = torch.from_numpy(state).float()
    probs, state_value = model(state)
    m = Categorical(probs)
    action = m.sample()
    model.saved_actions.append(SavedAction(m.log_prob(action), state_value))
    return action.item()
# In this function, we decide whehter we want the block to move left or right,based on what the model decided
```

```python
def finish_episode_old():
    # We calculate the losses and perform backprop in this function
    R = 0
    saved_actions = model.saved_actions
    policy_losses = []
    value_losses =[]
    returns = []
    
    for r in model.rewards[::-1]:
        R = r + 0.99 * R # 0.99 is our gamma number
        returns.insert(0, R)
    returns = torch.tensor(returns)
    returns = (returns - returns.mean()) / (returns.std() + eps)
    
    for (log_prob, value), R in zip(saved_actions, returns):
        advantage = R - value.item()
        
        policy_losses.append(-log_prob * advantage)
        value_losses.append(F.smooth_l1_loss(value, torch.tensor([R])))
    
    optimizer.zero_grad()
    loss = torch.stack(policy_losses).sum() + torch.stack(value_losses).sum()
    
    loss.backward()
    optimizer.step()
    
    del model.rewards[:]
    del model.saved_actions[:]
```

```python
def finish_episode():
    # We calculate the losses and perform backprop in this function
    global model
    R = 0
    saved_actions = model.saved_actions
    policy_losses = []
    value_losses =[]
    returns = []
    curr_return = []
    
    for r in model.rewards[::-1]:
        R = r + 0.99 * R # 0.99 is our gamma number
        returns.insert(0, R)
        curr_return.insert(0,r)

    returns = torch.tensor(returns)
    returns = (returns - returns.mean()) / (returns.std() + eps)
    
    cnt = 1
    for (log_prob, value),(next_log_prob, next_value ), R , r in zip(saved_actions[0:-1], saved_actions[1:], returns, model.rewards ):

        #advantage = R - value.item()
        Qsa = value.detach().item()
        td_error = r + 0.99*next_value - value

        # Compute the policy gradient
        #log_prob.backward(retain_graph=True)  # compute the gradient of the log probability
        #policy_gradient = td_error * torch.cat([p.grad() for p in model.parameters()])  # multiply by the TD error
        #ActorCritic.zero_grad()  # reset the gradients for the next update

        policy_losses.append(log_prob * td_error.detach())
        
        if cnt%5 == 0:
            delt = r + 0.99*next_value
            value_losses.append( F.mse_loss(value, delt.detach()))

        cnt += 1
    
    optimizer.zero_grad()
    loss = torch.stack(policy_losses).sum() + torch.stack(value_losses).sum()
    
    loss.backward()
    optimizer.step()
    
    del model.rewards[:]
    del model.saved_actions[:]
```

```python
old = False
def train():
    running_reward = 10
    for i_episode in count(): # We need around this much episodes
        state, info = env.reset()
        ep_reward = 0
        for t in range(1, 10000):
            action = select_action(state)
            state, reward, done, _ , _ = env.step(action)  #new_obs, reward, terminated, truncated, info
            model.rewards.append(reward)
            ep_reward += reward
            if done:
                break
        running_reward = 0.05 * ep_reward + (1-0.05) * running_reward
        if old:
            finish_episode_old()
        else:
            finish_episode()
        if i_episode % 10 == 0: # We will print some things out
            print("Episode {}\tLast Reward: {:.2f}\tAverage reward: {:.2f}".format(
                i_episode, ep_reward, running_reward
            ))
        if running_reward > env.spec.reward_threshold:
            print("Solved, running reward is now {} and the last episode runs to {} time steps".format(
                    running_reward, t
            ))
            break
            # This means that we solved cartpole and training is complete
```

```python
train()
```

```python
# There. we finished
# Lets see it in action
done = False
cnt = 0

env = gym.make('CartPole-v1', render_mode="human")

observation, info = env.reset()
while True:
    cnt += 1
    env.render()
    action = select_action(observation)
    observation, reward, done, _, _ = env.step(action)
    # Lets see how long it lasts until failing
    if done:
        break
print(f"Game lasted {cnt} moves")

env.close()
```
